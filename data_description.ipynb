```python
import pandas as pd

df = pd.read_csv('new.csv')
empty_percent = (df.isnull().sum().sum() / (df.shape[0] * df.shape[1])) * 100
print("Ilość wierszy", len(df))
print("Procent pustych danych", empty_percent)
df.describe()
```

Parametry: https://www.physionet.org/content/vitaldb/1.0.0/clinical_parameters.csv

```python
import matplotlib.pyplot as plt

column_names = df.columns
for column in column_names:
    plt.figure(figsize=(10, 10))
    df[column].hist()
    plt.title(column)
    plt.show()
```

```python
import matplotlib.pyplot as plt

df = pd.read_csv('colorectal_refactored.csv')
column_names = df.columns
for column in column_names:
    plt.figure(figsize=(10, 10))
    df.boxplot(column=column)
    plt.title(column)
    plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import colors

df['sex'].dropna()
df['age'].dropna()
df['optime'].dropna()
df['weight'].dropna()

plt.figure(figsize=(12, 12))
plt.scatter(df['age'], df['optime'], c=df['sex'], cmap=colors.ListedColormap(['red', 'green']), s=df['weight']*2, edgecolors='white')
```
