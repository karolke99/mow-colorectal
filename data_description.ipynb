```python
import matplotlib.pyplot as plt
from matplotlib import colors

df['sex'].dropna()
df['age'].dropna()
df['optime'].dropna()
df['weight'].dropna()

plt.figure(figsize=(12, 12))
plt.scatter(df['age'], df['optime'], c=df['sex'], cmap=colors.ListedColormap(['red', 'green']), s=df['weight']*2,edgecolors='white')

plt.scatter([], [], c='red', label='Male')
plt.scatter([], [], c='green', label='Female')
plt.scatter([], [], s=31*2, label='Weight: 48')
plt.scatter([], [], s=108*2, label='Weight: 108')


plt.xlabel('Age')
plt.ylabel('Operation time')
plt.legend()

plt.show()
```

### Opis wykresu

Wykres przedstawia zależność długości trwania operacji jelita grubego w zależności od wieku pacjenta. Zielonymi punktami oznaczono próbki pochodzące od kobiet, a czerwonymi - od mężczyzn. Wielkość punktu odpowiada wadze pacjenta. Im większy punkt, tym większa waga pacjenta.

Na podstawie wykresu można zauważyć, że dla starszych pacjentów czas operacji był dłuższy. Ponadto, w większości długie czasy operacji (>15k min) zaobserwowano dla mężczyzn. Istnieje kilka próbek, dla których pacjenci z większą wagą byli operowani dłużej. W grupie młodych osób większy odsetek krócej operowanych stanowili mężczyźni. Waga nie miała znaczącego wpływu na czas trwania operacji.
